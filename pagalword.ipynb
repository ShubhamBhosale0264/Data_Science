```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
data = requests.get('https://www.pagalworld.o   nline/top-songs.html')
```

```python
data
```

```python
data.content
```

```python
bdata = BeautifulSoup(data.content)
```

```python
bdata
```

```python
artist_name = bdata.find_all(class_="listbox-tags")
artist_name
```

```python
artistlist = []
for i in artist_name:
    artistlist.append(i.text)
artistlist
```

```python
songs = bdata.find_all('h2')
songs
```

```python
songlist = []
for i in songs:
    songlist.append(i.text)
songlist
```

```python
list_csv = pd.DataFrame({})
list_csv['Song_Name'] = songlist
list_csv['Artist_Name'] = artistlist
```

```python
list_csv.to_csv('Music_DataSet.csv')
```
