```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://www.magicbricks.com/property-for-sale/residential-real-estate?proptype=Multistorey-Apartment,Builder-Floor-Apartment,Penthouse,Studio-Apartment,Residential-House,Villa&cityName=Navi-Mumbai&category=B&parameter=rel&hideviewed=N&ListingsType=I&filterCount=3&incSrc=Y&fromSrc=homeSrc")
```

```python
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
Home_prize = soup.find_all(class_="mb-srp__card__price--amount")
Home_prize
```

```python
prize_list = []
for i in Home_prize:
    prize_list.append(i.text)
prize_list
```

```python
Home_Location = soup.find_all(class_="mb-srp__card--title")
Home_Location
```

```python
Location_list = []
for i in Home_Location:
    Location_list.append(i.text)
Location_list
```

```python
Status = soup.find_all(class_="mb-srp__card__summary--value")
Status
```

```python
Availability_list = []
for i in Status:
    Availability_list.append(i.text)
Availability_list
```

```python
MagicBricks = pd.DataFrame({})
MagicBricks["Prize"] = prize_list
MagicBricks["Location"] = Location_list
# MagicBricks['Status'] = Availability_list
MagicBricks
```

```python
MagicBricks.to_csv("Houses_in_NaviMumbai")
```
