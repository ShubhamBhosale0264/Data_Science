```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.arange(0, 100)
b = a * 2
c = a ** 2
```

```python
fig = plt.figure()
axs1 = fig.add_axes([0, 0, 1, 1])
axs2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
axs1.plot(a,b)
axs2.plot(a,b)
axs1.set_xlabel('X Label')
axs1.set_ylabel('Y Label')
axs1.set_title('Main Plot of a versus b')
axs2.set_xlabel('X Label')
axs2.set_ylabel('Y Label')
axs2.set_title('Inset Plot of x versus y')
plt.show()
```
