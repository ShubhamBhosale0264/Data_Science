```python
age = np.array([2,4,5,6,7,1,6,5,9,6])
speed = np.array([100,103,212,91,87,60,98,77,79,89])
plt.scatter(age,speed,cmap='hot')
plt.show()
```

```python
age = np.array([2,4,5,6,7,1,6,5,9,6])
speed = np.array([100,103,212,91,87,60,98,77,79,89])
plt.bar(age,speed,color='red',width=0.9)
plt.grid(color = 'black')
plt.xlabel("Age")
plt.ylabel("Speed")
plt.title("Age and Speeed Comparison")
plt.show()
```

```python
age = np.array([2,4,5,6,7,1,6,5,9,6])
# speed = np.array([100,103,212,91,87,60,98,77,79,89])
plt.hist(age,color='red',width=0.5)
plt.grid(color = 'black')
plt.xlabel("Age")
plt.ylabel("Speed")
plt.title("Age and Speeed Comparison")
plt.show()
```
