```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from matplotlib import pyplot as plt
```

```python
page = requests.get("https://groww.in/stocks/filter")
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content)
```

```python
stock_name = soup.find_all(class_ = "st76SymbolName")
stock_name
```

```python
stock_name_list = []
for i in stock_name:
    stock_name_list.append(i.text)
stock_name_list
```

```python
stock_prize = soup.find_all(class_ = "st76CurrVal bodyBaseHeavy")
stock_prize
```

```python
stock_prize_list = []
for i in stock_prize:
    stock_prize_list.append(i.text)
stock_prize_list
```

```python
stock_closing_prize = soup.find_all(class_ = "contentPrimary st76Pad16 bodyBase")
stock_closing_prize
```

```python
stock_closing_prize_list = []
for i in stock_closing_prize:
    stock_closing_prize_list.append(i.text)
stock_closing_prize_list
```

```python
market_cap = soup.find_all(class_="contentPrimary st76Pad16 bodyBase")
market_cap
```

```python
market_cap_list = []
for i in market_cap:
    market_cap_list.append(i.text)
market_cap_list
```

```python
stock_data =pd.DataFrame({})
stock_data["Stock_name"] =stock_name_list
stock_data["Stock_Prize"] =stock_prize_list
# stock_data["Stock_Closing_Prize"] =stock_closing_prize_list
stock_data["Market Capital(In CR)"] = market_cap_list?
stock_data
```

```python
a = stock_data['Stock_name']
b = stock_data['Stock_Prize']
plt.stem(a, b, label='Stock Data')  
plt.title("Relation between Stock Price and Stock Name", fontsize=14)
plt.xticks(rotation=90, ha='right', fontsize=10)  
plt.legend()
plt.show()
```
