## Checking unique doug crew

```python
unique_doug_crew = df[df['Crew'].str.contains("Doug", na=False)]['Crew'].unique()
```

```python
df['Crew'] = df.Crew.str.replace('Doug Wheellock Tracy Caldwell Dyson','Doug Wheellock       Tracy Caldwell Dyson')
```

## Counting unique doug crew

```python
df.Crew.value_counts()
```

```python
df['Crew'] = df['Crew'].str.replace('\s\s+', ',', regex=True)
crew_value_counts = df['Crew'].value_counts()
```

```python
df['Astronaut'] = df['Crew'].str.split(', ')
astronaut_set = set()
for crew_list in df['Astronaut']:
    astronaut_set.update(crew_list)
second_crew = [crew[1] if len(crew) > 1 else None for crew in df['Astronaut']]
astronaut_set.update(second_crew)
astronaut_df = pd.DataFrame(astronaut_set, columns=['Crew'])
astronaut_df['EVA_Number'] = 0
astronaut_df['EVA_Time'] = 0
print(astronaut_df.head())
```

```python
astronaut_df.dropna(how = 'any',inplace = True)
astronaut_df.head()
```

## Calculating EVA Statistics for Astronaut Crews

```python
for crew in astronaut_df['Crew']:
    total_eva_number = 0
    total_eva_time = 0
    for index, row in df.iterrows():
        if crew in row['Astronaut']:
            total_eva_number += 1
            total_eva_time += row['Duration']
        astronaut_df.loc[astronaut_df['Crew'] == crew, 'EVA_Number'] = total_eva_number
    astronaut_df.loc[astronaut_df['Crew'] == crew, 'EVA_Time'] = total_eva_time
print(astronaut_df.head())
```

## Visualizing EVA Statistics by Crew

```python
def graph_eva(column):
    astronaut_df.sort_values(by=column, ascending=False, inplace=True)
    plt.figure(figsize=(12, 8))
    sns.barplot(data=astronaut_df, x=column, y="Crew", palette='crest_r')
    plt.ylabel(None)
    plt.yticks(fontsize = 3)
    plt.show()
```

```python
graph_eva('EVA_Number')
```

```python
graph_eva('EVA_Time')
```
