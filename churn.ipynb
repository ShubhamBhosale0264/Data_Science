## Plotting

# Distribution of 'Total day minutes'

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['total day minutes'], bins=20, kde=True,color="red")
plt.title('Distribution of Total Day Minutes')
plt.xlabel('Total Day Minutes')
plt.ylabel('Frequency')
plt.show()
```

# Boxplot of 'Total eve charge' by 'Churn'

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='churn', y='total eve charge', data=df,color='orange')
plt.title('Boxplot of Total Eve Charge by Churn')
plt.xlabel('Churn')
plt.ylabel('Total Eve Charge')
plt.show()
```

# Visualization 3: Count plot of 'International plan'

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='international plan', data=df,color='purple')
plt.title('Count Plot of International Plan')
plt.xlabel('International Plan')
plt.ylabel('Count')
plt.show()
```

# Scatter plot of 'Total intl minutes' vs 'Total intl charge'

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total intl minutes', y='total intl charge', data=df,color = "black")
plt.title('Scatter Plot of Total Intl Minutes vs Total Intl Charge')
plt.xlabel('Total Intl Minutes')
plt.ylabel('Total Intl Charge')
plt.show()
```

# Scatterplot of numerical columns


```python
plt.scatter(df['total day minutes'], df['total eve minutes'], c=df['total night minutes'], cmap='viridis')
plt.title('Scatterplot of Total Day Minutes vs Total Eve Minutes')
plt.xlabel('Total Day Minutes')
plt.ylabel('Total Eve Minutes')
plt.colorbar(label='Total Night Minutes')
plt.show()
```

# Heatmap of correlation matrix

```python
numeric_df = df.select_dtypes(include='number')
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Accent')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

#  Boxplot of 'Total day charge' by 'Voice mail plan'


```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='voice mail plan', y='total day charge', data=df,palette=['red','yellow'])
plt.title('Boxplot of Total Day Charge by Voice Mail Plan')
plt.xlabel('Voice Mail Plan')
plt.ylabel('Total Day Charge')
plt.show()
```

# Count the number of churned and non-churned customers

```python
churn_counts = df['churn'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(churn_counts, labels=churn_counts.index,  autopct='%1.1f%%', colors=['darkorange', 'lightgreen'], explode=(0.1, 0), startangle=140)
plt.title('Churn Distribution')
plt.axis('equal')
plt.show()
```

# Stacked Total Minutes Distribution Across Customers

```python
total_day_minutes = df['total day minutes']
total_eve_minutes = df['total eve minutes']
total_night_minutes = df['total night minutes']
plt.figure(figsize=(10, 6))
plt.stackplot(range(len(df)), total_day_minutes, total_eve_minutes, total_night_minutes, labels=['Total Day Minutes', 'Total Eve Minutes', 'Total Night Minutes'], colors=['green', 'white', 'orange'])
plt.title('Stackplot of Total Minutes')
plt.xlabel('Customer')
plt.ylabel('Total Minutes')
plt.legend(loc='upper left')
plt.show()
```

# Violin plot for 'Total night calls'

```python
plt.subplot(2, 20, 4)
sns.violinplot(y='total night calls', data=df, color='blue')
plt.title('Violin Plot of Total Night Calls')
```
